# 🚢 Titanic Dataset: Measures of Central Tendency (Mean, Median, Mode)

In this notebook, we explore the fundamental statistical concepts of central tendency using the famous Titanic passengers dataset.

## 1. Theoretical Concepts
In statistics, we use three main measures to describe the "center" of a data distribution:

> Mean: The arithmetic average, calculated by summing all values and dividing by the total count.

> Formula:$$\bar{x} = \frac{\sum x_i}{n}$$Median: The middle value when the data is sorted in ascending order. It is more robust to outliers than the mean.

> Mode: The value that appears most frequently in the dataset. A dataset can have one mode, multiple modes, or no mode at all.

## 2. Loading Libraries and Data

```python
import pandas as pd
import seaborn as sns

# Load the Titanic dataset
df = sns.load_dataset('titanic')

# Display the first 5 rows
df.head()
```

## 3. Practical Analysis: Passenger Age (age)
Let's analyze the age distribution of the passengers on board.

### A) Arithmetic Mean
What was the average age of the Titanic passengers?

```python
mean_age = df['age'].mean()
print(f"Mean Age: {mean_age:.2f} years")
```

### B) Median
Finding the central point of the ages. This helps us understand the age of the "middle" passenger.

```python
median_age = df['age'].median()
print(f"Median Age: {median_age} years")
```

### C) Mode
Which specific age was the most common among passengers?

```python
mode_age = df['age'].mode()[0]
print(f"Most Frequent Age (Mode): {mode_age} years")
```

## 4. Summary Table

| Metric | Value | Description |
| :--- | :---: | :--- |
| **Mean** | `mean_age` | The average age of all passengers. |
| **Median** | `median_age` | The exact middle age in the distribution. |
| **Mode** | `mode_age` | The most frequent age observed. |

## 5. Conclusion

By comparing these three values, we can see if the age distribution is symmetrical or skewed. For instance, if the Mean is significantly higher than the Median, it suggests there were a few very elderly passengers pulling the average up.
